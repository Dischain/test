```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

from mpl_toolkits.mplot3d import Axes3D

X = pd.read_csv("./data/train_values.csv")
y = pd.read_csv("./data/train_labels.csv")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.sort_index(inplace=True)
X_test.sort_index(inplace=True)
y_train.sort_index(inplace=True)
y_test.sort_index(inplace=True)

data = pd.DataFrame()
data = pd.merge(X_train, y_train, on='building_id', how='inner')

categorical_cols = ['land_surface_condition', 'foundation_type', 'roof_type',
                    'ground_floor_type', 'other_floor_type', 'position',
                    'plan_configuration', 'legal_ownership_status']

binary_cols_superstructure = ['has_superstructure_adobe_mud', 'has_superstructure_mud_mortar_stone', 
                              'has_superstructure_stone_flag', 'has_superstructure_cement_mortar_stone',
                              'has_superstructure_mud_mortar_brick', 'has_superstructure_cement_mortar_brick',
                              'has_superstructure_timber', 'has_superstructure_bamboo',
                              'has_superstructure_rc_non_engineered', 'has_superstructure_rc_engineered',
                              'has_superstructure_other']

binary_cols_secondary_use = ['has_secondary_use',
                             'has_secondary_use_agriculture', 'has_secondary_use_hotel',
                             'has_secondary_use_rental', 'has_secondary_use_institution',
                             'has_secondary_use_school', 'has_secondary_use_industry',
                             'has_secondary_use_health_post', 'has_secondary_use_gov_office',
                             'has_secondary_use_use_police', 'has_secondary_use_other']

numeric_cols = ['count_floors_pre_eq', 'age', 'area_percentage', 'height_percentage', 'count_families']

descrete_cols = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id']
```

### 1. Numerical features analisys

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
for variable, subplot in zip(numeric_cols, ax.flatten()):
    sns.countplot(data[variable], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
```

```python
sns.pairplot(data[numeric_cols])
```

```python
corr_m = data[numeric_cols].corr()

mask = np.zeros_like(corr_m, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_m, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### 1.2 Count floors before earthquake

```python
legend = []

for i in range(data['count_floors_pre_eq'].max()):
    floor = i + 1
    data[data['count_floors_pre_eq'] == floor].groupby('damage_grade').count()['count_floors_pre_eq'].plot()
    legend.append(floor)
    
plt.legend(legend)
```

```python
gl_X_train = X_train[descrete_cols]
gl_y_train = y_train[['damage_grade']]

gl_X_test = X_test[descrete_cols]
gl_y_test = y_test[['damage_grade']]
```

```python
gl_pca = PCA(n_components=2)
gl_pca.fit(gl_X_train - gl_X_train.mean(axis=0))

fitted_gl_pca = pd.DataFrame()

fitted_gl_pca = pd.DataFrame(gl_pca.transform(gl_X_train - gl_X_train.mean(axis=0)))
```

```python
gl_KM = KMeans(n_clusters=15)
gl_KM.fit(gl_X_train)
gl_predicted = pd.DataFrame(gl_KM.predict(gl_X_train))

gl_predicted['building_id'] = X_train['building_id'].values
gl_predicted = gl_predicted.merge(y_train, on='building_id')
```

```python
ax1 = sns.distplot(gl_predicted[gl_predicted['damage_grade'] == 1][0])
```

```python
ax3 = sns.distplot(gl_predicted[gl_predicted['damage_grade'] == 3][0])
```

```python
gl_KM.score(gl_X_train)
```

```python
gl_RFC = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
gl_RFC.fit(gl_X_train, gl_y_train)

gl_RFC.score(gl_X_train, gl_y_train)
```

```python
gl_knn_pipe = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_jobs=-1))])

gl_knn_params = {'knn__n_neighbors': range(1, 20)}

gl_knn_grid = GridSearchCV(gl_knn_pipe, gl_knn_params, cv=5, n_jobs=-1, verbose=True)

gl_knn_grid.fit(gl_X_train, gl_y_train)

gl_knn_grid.best_params_, gl_knn_grid.best_score_

# ({'knn__n_neighbors': 15}, 0.7031305483327797)
```

```python
gl_knn = KNeighborsClassifier(n_neighbors=11)
gl_knn.fit(gl_X_train, gl_y_train)
gl_knn.score(gl_X_test, gl_y_test)
# 0.6936824846800544 - 9
# 0.6938917894394121 - 15
# 0.6956825079361388 - 11
```
